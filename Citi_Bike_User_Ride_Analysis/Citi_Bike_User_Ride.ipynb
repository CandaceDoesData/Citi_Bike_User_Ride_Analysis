```python
# Import dependencies
import pandas as pd
import os
from datetime import timedelta
```

```python
# Define the directory where CSV files are located
csv_directory = "/Users/candacestingley/Documents/Bootcamp_Folder/Module_18/Module_18_Challenge/Starting_Over/raw_csv"
```

```python
# Create empty list to store the dataframes of each CSV file
dataframes = []
```

```python
# Read in the CSV files and append their dataframes to the list
for filename in os.listdir(csv_directory):
    if filename.endswith(".csv"):
        filepath = os.path.join(csv_directory, filename)
        df = pd.read_csv(filepath)
        dataframes.append(df)
```

```python
# Combine the dataframes into one long dataframe
new_19_20_df = pd.concat(dataframes, ignore_index=True)
```

```python
new_19_20_df.head()
```

```python
# Check column data types
new_19_20_df.dtypes
```

```python
new_19_20_df = new_19_20_df.rename(columns = {'tripduration':'trip_seconds',
                                              'starttime':'start_time',
                                              'stoptime':'end_time',
                                              'start station id':'start_id',
                                              'start station name':'start_name',
                                              'start station latitude':'start_lat',
                                              'start station longitude':'start_long',
                                              'end station id':'end_id',
                                              'end station name':'end_name',
                                              'end station latitude':'end_lat',
                                              'end station longitude':'end_long',
                                              'bikeid':'bike_id',
                                              'usertype': 'user_type',
                                              'birth year':'birth_year'})
```

```python
# Create a new column that calculates the user's age based off of "birth_year"
current_year = 2023
new_19_20_df['user_age'] = current_year - new_19_20_df['birth_year']
new_19_20_df.head()
```

```python
# Convert "start_time" and "end_time" columns to datetime dtype
new_19_20_df['start_time'] = pd.to_datetime(new_19_20_df['start_time'])
new_19_20_df['end_time'] = pd.to_datetime(new_19_20_df['end_time'])
```

```python
new_19_20_df = new_19_20_df[['trip_seconds',
                             'start_time','end_time',
                            'start_id','start_name',
                            'start_lat','start_long',
                            'end_id','end_name',
                            'end_lat','end_long',
                            'user_type','user_age',
                            'birth_year','gender',
                            'bike_id']]
```

```python
new_19_20_df.head()
```

```python
# Change gender column and datatype from integers to strings
new_19_20_df['gender'] = new_19_20_df['gender'].replace({0:'Unknown', 1:'Male', 2:'Female'})
new_19_20_df['gender'] = new_19_20_df['gender'].astype('object')
```

```python
new_19_20_df.dtypes
```

```python
# Save the dataframe as a CSV file
output_csv_path = "/Users/candacestingley/Documents/Bootcamp_Folder/Module_18/Module_18_Challenge/Starting_Over/new_19_20_df.csv"

new_19_20_df.to_csv(output_csv_path, index=False)
```

```python
station_count = len(new_19_20_df)
station_count
```

```python
start_count_df = new_19_20_df['start_name'].value_counts()
start_count_df = pd.DataFrame(start_count_df)
start_count_df = start_count_df.rename(columns = {'start_name':'start_count'})
start_count_df = start_count_df.rename_axis('station_name', axis ='index').reset_index()
start_count_df = start_count_df.sort_values(by = 'station_name')
start_count_df.head()
```

```python
end_count_df = new_19_20_df['end_name'].value_counts()
end_count_df = pd.DataFrame(end_count_df)
end_count_df = end_count_df.rename(columns = {'end_name':'end_count'})
end_count_df = end_count_df.rename_axis('station_name', axis ='index').reset_index()
end_count_df = end_count_df.sort_values(by = 'station_name')
end_count_df.head()
```

```python
start_end_count_df = pd.merge(end_count_df, start_count_df, on = 'station_name', how = 'outer')
start_end_count_df = start_end_count_df[['station_name','start_count','end_count']]
start_end_count_df['start_count'].fillna(0, inplace = True)
start_end_count_df['end_count'].fillna(0, inplace = True)
start_end_count_df.head()
```

```python
start_end_count_df['start_count'] = start_end_count_df['start_count'].astype(int)
```

```python
start_end_count_df['total_count'] = start_end_count_df['start_count'] + start_end_count_df['end_count']
start_end_count_df.head()
```

```python
start_end_count_df['start_percentage'] = (start_end_count_df['start_count'] / start_end_count_df['total_count'])
start_end_count_df['start_percentage'] = start_end_count_df['start_percentage'].round(2)
```

```python
start_end_count_df['end_percentage'] = (start_end_count_df['end_count'] / start_end_count_df['total_count'])
start_end_count_df['end_percentage'] = start_end_count_df['end_percentage'].round(2)
```

```python
start_end_count_df.head(10)
```

```python
start_end_count_df['station_percentage'] = (start_end_count_df['total_count'] / station_count) 
start_end_count_df['station_percentage'] = start_end_count_df['station_percentage'].round(4)
start_end_count_df.head()
```

```python
# Save the dataframe as a CSV file
output_csv_path = "/Users/candacestingley/Documents/Bootcamp_Folder/Module_18/Module_18_Challenge/Starting_Over/start_end_count_df.csv"

start_end_count_df.to_csv(output_csv_path, index=False)
```
